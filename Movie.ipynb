```python
movies=final_df[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
import ast
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies
```

```python
def convert3(obj):
  l=[]
  count=0
  for i in ast.literal_eval(obj):
    if count !=3:
      l.append(i['name'])
      count+= 1
    else:
      break
  return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fe(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(fe)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['overview']=movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies
```

```python
movies['tags']=movies['cast']+movies['crew']+movies['overview']+movies['keywords']+movies['genres']
```

```python
movies['tags']
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vecrot=cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi=cosine_similarity(vecrot)
```

```python
from scipy.spatial import distance
def recommend(movie):
  movie_index=new_df[new_df['title'] == movie].index[0]
  distance = simi[movie_index]
  movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:11]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
    cast=(movies.iloc[i[0]].cast)
    text = ' '.join(cast)
    print(text)




```

```python
recommend('Avatar')
```
